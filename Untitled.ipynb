```python
# 在字符串full_text中，通过discourse_start,discourse_end计算predictionstring
def calc_word_indices(full_text, discourse_start, discourse_end):
    start_index = len(full_text[:discourse_start].split())
    token_len = len(full_text[discourse_start:discourse_end].split())
    output = list(range(start_index, start_index + token_len))
    if output[-1] >= len(full_text.split()):
        output = list(range(start_index, start_index + token_len-1))
    return output
```

```python
type_to_id = {
    'Lead':0, 
    'Position':1, 
    'Evidence':2, 
    'Claim':3, 
    'Concluding Statement':4,
    'Counterclaim':5, 
    'Rebuttal':6
}
```

```python
tokenizer = AutoTokenizer.from_pretrained("allenai/longformer-base-4096")
tokenizer.add_prefix_space = True
```

```python
df = pd.read_csv("input/feedback-prize-2021/train.csv")
train_folder_name = "input/feedback-prize-2021/train/"
prediction_string_starts = df['discourse_start']
prediction_string_ends = df['discourse_end']

ids = df['id'].unique()
df.head()
```

```python
train_samples = {}
train_samples_length = []
train_texts_length = []
```

```python
index = 0
for ids_id in ids:
  with open(train_folder_name + ids_id + '.txt') as file:
      text = file.read()
  encoded_text = tokenizer.encode_plus(text.split(),
                                       padding='max_length',
                                       is_split_into_words=True,
                                       add_special_tokens=True, 
                                       return_offsets_mapping=True,
                                       return_tensors="tf")
  train_samples[ids_id] = {
      'index': index,
      'text' : text,
      'encoded_text': encoded_text
  }
  train_texts_length.append(len(text))
  train_samples_length.append(encoded_text.input_ids.shape[1])
  file.close()
  index = index + 1
targets = np.zeros([len(train_samples), np.max(train_samples_length), 15 ], dtype='int32')
```

```python
g = df.groupby(df['id'])
for i in ids:
    gg = g.get_group(i)
    for s, e, t in zip(gg['discourse_start'], 
                        gg['discourse_end'],
                        gg['discourse_type']
                       ):
        xs = calc_word_indices(train_samples[i]['text'], int(s), int(e))
        for x in xs:  
            if x > 0 :
                if targets[i][x-1,type_to_id[t]] == 1 or targets[i][x-1,type_to_id[t] + 7] == 1:
                    targets[i][x,type_to_id[t] + 7] = 1
                else:
                    targets[i][x,type_to_id[t]] = 1
            else:
                targets[i][x,type_to_id[t]] = 1
```
